```python
import pandas as pd
import numpy as np
import plotly.express as ax
```

```python
# pd.pandas.set_option("display.max_columns", None)
File = open("data.txt", 'r')
```

```python
Lines = File.readlines()
p=[]
w=[]
t=[]
for line in Lines:
    w.append(int(line.split()[0]))
    p.append(float(line.split()[1]))
    t.append(int(line.split()[2]))
```

```python
numpy_array = np.array((p, w, t), dtype = float)
dataset = pd.DataFrame({'Probability':p, 'Width':w, 'Time':t})

import plotly.express as px

fig = px.scatter_3d(dataset, x = 'Probability', y = 'Width', z = 'Time', color='Time' )
fig.show()
```
